```python
import pandas as pd
```

```python
# data 출처: https://www.kaggle.com/hesh97/titanicdataset-traincsv/data
train_data = pd.read_csv('../data/train.csv')
train_data.head()
```

#### dataframe slicing
  - dataframe의 경우 기본적으로 [] 연산자가 **column 선택**에 사용
  - 하지만, slicing은 row 레벨로 지원

```python
train_data[:10] # 10 미만까지
# slicing만 예외!! []로 index접근 가능
```

#### row 선택하기
  - Seires의 경우 []로 row 선택이 가능하나, **DataFrame의 경우는 기본적으로 column을 선택하도록 설계**
  - **.loc, .iloc함수**로 row 선택 가능
    - loc - 인덱스 자체를 사용
    - iloc - 0 based index로 사용
    - 이 두 함수는 ,를 사용하여 column 선택도 가능

```python
train_data.head()
```

```python
train_data.info()
```

```python
import numpy as np
train_data.index = np.arange(100, 991)
train_data.head()
```

```python
train_data.loc[986] # Series 형태
```

```python
train_data.loc[[986, 100, 101, 110]] # multiple한 index : list로
```

```python
train_data.iloc[0] # 어떤 index이든지 간에 순서대로 가지고 오고 싶을 때
```

```python
train_data.iloc[[0, 100, 101, 110]]
```

### row, column 동시에 선택하기
- loc, iloc 속성을 이용할 때, 콤마를 이용하여 둘 다 명시 가능

```python
train_data.loc[[984, 100, 101, 990], ['Survived', 'Name', 'Sex', 'Age']]
```

```python
train_data.iloc[[0, 100, 101, 110], [1, 4, 5]]
```

```python
train_data.loc[[984, 100, 101, 990], [1, 4, 5]]
```
