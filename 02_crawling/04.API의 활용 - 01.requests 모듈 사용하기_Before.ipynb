### 학습목표 
 1. requests 모듈 사용하여 http request/response 확인하기

#### requests 모듈
1. http request/response를 위한 모듈
2. HTTP method를 메소드 명으로 사용하여 request 요청 예) get, post

```python
import requests
```

#### get 요청하기
1. http get 요청하기
2. query parameter 이용하여 데이터 전달하기

```python
url = 'https://news.v.daum.net/v/20190728165812603'
resp = requests.get(url) # 2로 시작하면 성공임

resp
resp.text # 어떤 data가 있는지 .text로 확인
```

#### post 요청하기
1. http post 요청하기
2. post data 이용하여 데이터 전달하기

```python
# 기본적인 page에서는 사용 x, 민감한 정보에서 확인 (주로 로그인 상황)

url = 'https://www.kangcom.com/member/login.asp'
data = {
    'id':'macmath22',
    'pwd':'Test1357!'
}
resp = requests.post(url, data=data) # post함수에서는 data 파라미터 존재
resp.text
```

#### HTTP header 데이터 이용하기
1. header 데이터 구성하기
2. header 데이터 전달하기

```python
# crawling이 잘 안되는 경우 : 기본적으로 headers탭에서 user-agent를 많이 씀
url = 'https://www.kangcom.com/member/login.asp'
headers = {
    'user-agent' : 'Mozilla/5.0 (Macintosh; Intel Mac OS X 11_2_1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/88.0.4324.192 Safari/537.36'
}

resp = requests.get(url, headers=headers)
resp.text
```

#### HTTP response 처리하기
1. response 객체의 이해
2. status_code 확인하기
3. text 속성 확인하기

```python
url = 'https://news.v.daum.net/v/20190728165812603'

resp = requests.get(url)
if resp.status_code == 200 :
    resp.text
else:
    print('error')
    
    
print(resp.status_code)
```
